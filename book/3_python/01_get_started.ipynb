## Using Colab

## Using Binder 

**Binder** (slower alternative) is another free online service that lets you run Jupyter notebooks in the cloud without local installation.

Binder launches a temporary Jupyter environment based on the course GitHub repository. Startup can take a minute or two, especially when many users access Binder at the same time.

The 🚀 *launch button* in the top right corner starts a temporary Binder session, when you copy `https://mybinder.org` as the URL placeholder. Once launched you can run the notebook in your browser.

> ⚠️ Binder sessions are temporary
> Any changes you make will be lost when the session ends unless you download your notebook.


---

## In-page execution 

This Jupyter Book supports **in-page execution** (temporary cloud kernel), which allows you to run code cells **directly on this page**, without opening a separate notebook interface.

When you click the power button <img src="./images/1_power-button.png" alt="rocket" style="height:1.2em; vertical-align:middle;"> (top of the page), a **temporary cloud-based Jupyter kernel** is started in the background. Once the session is ready, you can execute the code cells on this page and view the outputs inline.

When you click the power button ![power](./images/1_power-button.png) (top of the page), a **temporary cloud-based Jupyter kernel** is started in the background. Once the session is ready, you can execute the code cells on this page and view the outputs inline.

> 💡 This option is useful for **quick experiments, demonstrations, or checking understanding**.
> It is not intended for longer work sessions or saving progress.

> ⚠️ In-page execution uses a **temporary Binder-backed environment**
> Startup may take some time, and sessions are not persistent. Any changes or results are lost when the page is refreshed.

## Running the notebook locally

If you prefer to run notebooks locally on your own machine, you need a **local Python environment** and **JupyterLab**.
For this course, this means a Conda-based setup. See the Setup sections on **Conda** and **JupyterLab** for details.

Start JupyterLab from your environment:

```bash

```

This opens JupyterLab in your web browser. From there, navigate to the notebook file and run it locally.

## Running Locally

To run Jupyter notebooks locally on your machine, you need to have JupyterLab installed. You can install it using the following command:

```bash
pip install jupyterlab
```

Once you have JupyterLab installed, you can run it by executing the following command in your terminal:

```bash
jupyter lab
```

This will open JupyterLab in your default web browser, where you can navigate to the notebook you want to run.
